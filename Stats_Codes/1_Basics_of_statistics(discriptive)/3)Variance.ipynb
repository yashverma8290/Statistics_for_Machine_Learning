```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
class_A = [85, 87, 88, 86, 84]
class_B = [60, 90, 55, 95, 50]

# Compute Variance
var_A = np.var(class_A, ddof=0)  # Population Variance
var_B = np.var(class_B, ddof=0)  # Population Variance

print("Variance of Class A:", var_A)
print("Variance of Class B:", var_B)

# Visualization
plt.figure(figsize=(10, 5))
sns.histplot(class_A, bins=5, kde=True, color="blue", label="Class A (Low Variance)")
sns.histplot(class_B, bins=5, kde=True, color="red", label="Class B (High Variance)")
plt.axvline(np.mean(class_A), color='blue', linestyle='dashed', label="Mean of Class A")
plt.axvline(np.mean(class_B), color='red', linestyle='dashed', label="Mean of Class B")
plt.legend()
plt.title("Comparison of Variance")
plt.xlabel("Scores")
plt.ylabel("Density")
plt.show()
```

# Explanation

This visualization represents the concept of variance by comparing the distribution of scores from two different classes, Class A and Class B. Variance is a statistical measure that indicates how far data points deviate from the mean. A higher variance means the data points are more spread out, while a lower variance suggests they are closer to the mean. In this case, the variance of Class A is 2, meaning its values are tightly packed, whereas the variance of Class B is 350, showing a much wider spread.

The histogram and kernel density estimation (KDE) curves help illustrate these differences. The blue histogram and KDE curve represent Class A, which has a narrow distribution, signifying low variance. The red histogram and KDE curve represent Class B, where the scores are widely spread, resulting in high variance. This difference in spread is visually apparent, as the red bars cover a much broader range of values compared to the blue ones.

The mean of each class is also indicated with dashed lines. The blue dashed line represents the mean of Class A, which is closely surrounded by data points, reinforcing its low variance. On the other hand, the red dashed line represents the mean of Class B, but the data points are more scattered around it, confirming its high variance. The KDE curve for Class A is more concentrated and steep, indicating that most values are close to the mean. Meanwhile, the KDE curve for Class B is more spread out and flatter, emphasizing greater variation among its values.

From this visualization, it is evident that Class A has a more consistent distribution of scores, while Class B has a greater range of values. Variance is an essential statistical measure used to understand data dispersion, helping in various fields like finance, machine learning, and scientific research. A high variance could indicate unpredictability, while a low variance suggests more stability.
