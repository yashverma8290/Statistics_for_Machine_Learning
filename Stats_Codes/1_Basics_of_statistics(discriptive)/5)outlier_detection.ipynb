# OUTLIER DETECTION

# 1) IQR Method

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
scores = [10, 20, 25, 30, 35, 40, 45, 50, 55, 90]

# Calculate Quartiles
Q1 = np.percentile(scores, 25)  # First Quartile (25th percentile)
Q3 = np.percentile(scores, 75)  # Third Quartile (75th percentile)

# Calculate IQR
IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")

# Visualization using a boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(y=scores)
plt.title("Boxplot of Scores (IQR Visualization)")
plt.ylabel("Scores")
plt.show()
```

# Explanation 
The boxplot shown in the image represents the distribution of a dataset using quartiles and the interquartile range (IQR). The first quartile (Q1), which is the 25th percentile, is 26.25, meaning that 25% of the data points fall below this value. The third quartile (Q3), the 75th percentile, is 48.75, meaning that 75% of the data lies below this value. The interquartile range (IQR) is calculated as Q3 - Q1 = 48.75 - 26.25 = 22.5, representing the spread of the middle 50% of the data.

The box in the boxplot represents the IQR, where the horizontal line inside the box indicates the median, the middle value of the dataset. The whiskers, which extend from both ends of the box, show the range of the data, except for any outliers. In this case, an outlier is observed at 90, which appears as a separate dot above the box. Outliers are identified using the rule that any value below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR is considered an outlier. Here, the upper bound is 82.5, and since 90 exceeds this value, it is flagged as an outlier.

Overall, the boxplot visually represents how the majority of the data points are concentrated within the interquartile range, while a few extreme values, like 90, deviate significantly from the rest of the dataset. This makes IQR a useful measure for detecting data variability and identifying potential outliers in a dataset.










# 2) Z-Score

```python
import numpy as np
from scipy.stats import zscore

# Example dataset (exam scores)
data = np.array([60, 70, 80, 90, 95, 85, 75, 65])

# Calculate Z-Scores
z_scores = zscore(data)

# Print results
for i, z in enumerate(z_scores):
    print(f"Value: {data[i]}, Z-Score: {z:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore, norm

# Generate a dataset (random normal distribution)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=1000)  # Mean = 70, Std Dev = 10

# Calculate Z-scores
z_scores = zscore(data)

# Create the plot
plt.figure(figsize=(10, 5))
sns.histplot(z_scores, bins=30, kde=True, color='blue', stat="density", edgecolor='black')

# Generate a standard normal distribution curve
x = np.linspace(-3, 3, 1000)  # Z-score range (-3 to 3)
y = norm.pdf(x, 0, 1)  # Standard normal distribution (mean=0, std=1)
plt.plot(x, y, 'r-', label="Standard Normal Curve")

# Mark key Z-score positions
for z in [-2, -1, 0, 1, 2]:
    plt.axvline(z, color='black', linestyle='dashed', alpha=0.7)

# Labels and title
plt.title("Z-Score Distribution with Standard Normal Curve")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.legend()
plt.show()
```

A Z-score, also known as a standard score, measures how far a particular data point is from the mean in terms of standard deviations. It helps in comparing values from different datasets by standardizing them onto a common scale. The Z-score is calculated using of z score. A Z-score of 0 indicates that the value is exactly at the mean, while positive and negative Z-scores represent values above and below the mean, respectively. In machine learning and artificial intelligence, Z-scores play a crucial role in outlier detection, feature scaling, and hypothesis testing. For instance, in fraud detection, extremely high or low Z-scores can signal anomalies. Additionally, many ML algorithms, such as logistic regression and KNN, benefit from standardized data, ensuring better performance and faster convergence. By transforming raw data into Z-scores, we can make datasets more interpretable and suitable for statistical analysis and machine learning models.
