# Example 
This example demonstrates a Discrete Uniform Distribution using the classic scenario of rolling a fair six-sided die. In a discrete uniform distribution, every outcome has an equal probability of occurring. Here, the die has possible outcomes of 1 to 6, and each face has a probability of 1/6 (since the die is fair).

We simulate the process of rolling the die 1000 times using random sampling. The simulated results are stored as an array representing the outcomes of all the rolls. After the simulation, we visualize the distribution using a histogram, where the x-axis shows the die faces (1 to 6) and the y-axis shows the relative frequencies (probabilities) of each face appearing based on the simulation.

In the plot, you will observe that each die face appears approximately equally often, confirming the characteristic of the uniform distribution. The histogram bars should be roughly the same height, indicating that each face is equally likely.

This type of distribution is commonly used in games of chance, randomized experiments, or any system where all outcomes are equally likely. It helps in understanding how probabilities are distributed equally in simple random experiments like rolling dice, drawing cards from a well-shuffled deck, or selecting random numbers from a fixed set. This experiment visually shows the foundational idea behind the discrete uniform distribution, which is essential in probability theory and basic statistics.

```python
import numpy as np
import matplotlib.pyplot as plt

# Possible outcomes of a fair die
values = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.ones(6) / 6  # Each outcome has probability 1/6

# Simulate rolling the die 1000 times
rolls = np.random.choice(values, size=1000, p=probabilities)

# Plot the histogram
plt.figure(figsize=(7,5))
plt.hist(rolls, bins=np.arange(0.5, 7.5, 1), density=True, rwidth=0.8, color="skyblue", edgecolor="black")

plt.xticks(values)
plt.title("Discrete Uniform Distribution (Rolling a Fair Die)")
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.show()
```

# Explanation 
This histogram represents a Discrete Uniform Distribution obtained by simulating the rolling of a fair six-sided die 1000 times. In this distribution, each possible outcome—die faces 1 to 6—has an equal theoretical probability of 1/6 (≈ 0.1667). The histogram visually demonstrates how often each face appears in the simulated rolls. Since the die is fair, the bars in the histogram are approximately equal in height, confirming that all outcomes occur with similar frequency. However, minor variations are present due to randomness in the finite sample size. Some outcomes appear slightly more or less frequently than expected, but as the number of rolls increases, the distribution will more closely match the theoretical probability. This experiment effectively illustrates the concept of a Discrete Uniform Distribution, which is widely used in probability theory, statistical simulations, and real-world applications such as games of chance and randomized decision-making.
