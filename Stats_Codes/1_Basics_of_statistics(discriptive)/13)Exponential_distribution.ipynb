```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import expon

# Define the rate parameter (lambda)
lambda_ = 2  # Average 2 calls per minute

# Generate 1000 random samples from Exponential Distribution
data = np.random.exponential(scale=1/lambda_, size=1000)

# Plot histogram
plt.figure(figsize=(8,5))
sns.histplot(data, bins=30, kde=True, color="blue", stat="density")

# Plot theoretical exponential PDF
x = np.linspace(0, max(data), 100)
pdf_values = expon.pdf(x, scale=1/lambda_)
plt.plot(x, pdf_values, 'r-', lw=2, label="Theoretical Exponential Curve")

# Labels and title
plt.title("Exponential Distribution of Call Waiting Times")
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("Density")
plt.legend()
plt.show()
```
