# Skewness 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skewnorm, mode

# Function to plot skewness
def plot_skewness(ax, data, title):
    sns.histplot(data, bins=30, kde=True, color='blue', ax=ax)
    ax.set_title(title)
    
    # Calculate statistics
    mean = np.mean(data)
    median = np.median(data)
    mode_val = mode(data, keepdims=True).mode[0]
    
    # Plot mean, median, and mode
    ax.axvline(mean, color='red', linestyle='dashed', label='Mean')
    ax.axvline(median, color='green', linestyle='dashed', label='Median')
    ax.axvline(mode_val, color='purple', linestyle='dashed', label='Mode')
    ax.legend()

# Generate datasets for different skewness types
np.random.seed(42)

# Negatively skewed data (left skewed)
neg_skew_data = skewnorm.rvs(a=-6, size=1000)

# Symmetric data (normal distribution)
sym_data = np.random.normal(loc=0, scale=1, size=1000)

# Positively skewed data (right skewed)
pos_skew_data = skewnorm.rvs(a=6, size=1000)

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
plot_skewness(axes[0], neg_skew_data, "Negatively Skewed (Left Skewed)")
plot_skewness(axes[1], sym_data, "Symmetric (Normal Distribution)")
plot_skewness(axes[2], pos_skew_data, "Positively Skewed (Right Skewed)")

plt.tight_layout()
plt.show()
```

Skewness is a statistical measure that describes the asymmetry of a datasets distribution. It indicates whether the data is skewed towards the left (negative skew), right (positive skew), or is symmetric (normal distribution). In a perfectly normal distribution, the mean, median, and mode are equal and lie at the center. However, in skewed distributions, these measures of central tendency differ.

A negatively skewed distribution (left-skewed) has a longer tail on the left side, meaning that most of the data values are concentrated on the right. In such cases, the mean is typically less than the median, and the mode appears at the highest peak of the distribution. An example of this could be exam scores where a few students perform very poorly while most achieve higher marks.

A positively skewed distribution (right-skewed) has a longer tail on the right side, indicating that most values are concentrated on the left. Here, the mean is greater than the median, and the mode appears at the highest peak. A real-world example is income distribution, where a majority earn lower wages, while a few individuals earn significantly higher salaries, pulling the mean upward.

A symmetric distribution (normal distribution) has no skewness, meaning the left and right sides are mirror images of each other. The mean, median, and mode coincide at the center of the distribution. A common example is height distribution in a population, where most peoples heights cluster around the average value.

In the provided code, three distributions are visualized: left-skewed, symmetric, and right-skewed. The red, green, and purple dashed lines indicate the mean, median, and mode, respectively. These visualizations help in understanding how skewness affects the spread of data and why understanding skewness is crucial for data analysis and machine learning, where assuming normality may not always be valid.

# Kurtosis

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Generate data for different types of kurtosis
np.random.seed(42)

# Mesokurtic (Normal Distribution, Kurtosis  3)
mesokurtic_data = np.random.normal(loc=0, scale=1, size=1000)

# Leptokurtic (High Kurtosis, Heavy Tails)
leptokurtic_data = np.random.laplace(loc=0, scale=1, size=1000)

# Platykurtic (Low Kurtosis, Light Tails)
platykurtic_data = np.random.uniform(low=-2, high=2, size=1000)

# Function to plot kurtosis
def plot_kurtosis(ax, data, title):
    sns.histplot(data, bins=30, kde=True, color='blue', ax=ax)
    ax.set_title(title)
    ax.axvline(np.mean(data), color='red', linestyle='dashed', label='Mean')
    ax.legend()

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
plot_kurtosis(axes[0], leptokurtic_data, "Leptokurtic (Heavy Tails, Kurtosis > 3)")
plot_kurtosis(axes[1], mesokurtic_data, "Mesokurtic (Normal, Kurtosis  3)")
plot_kurtosis(axes[2], platykurtic_data, "Platykurtic (Light Tails, Kurtosis < 3)")

plt.tight_layout()
plt.show()
```

Kurtosis is a statistical measure that describes the shape of a distribution, particularly its tails and peak compared to a normal distribution. It helps in understanding how extreme values (outliers) influence the distribution. There are three types of kurtosis: Mesokurtic, Leptokurtic, and Platykurtic. A Mesokurtic distribution has a kurtosis close to 3, resembling a normal distribution with a balanced peak and tails. An example is the distribution of human heights in a population. A Leptokurtic distribution has a kurtosis greater than 3, indicating a sharp peak and heavy tails, meaning more extreme values or outliers exist. A common example is financial market returns, where extreme gains or losses occur. In contrast, a Platykurtic distribution has a kurtosis less than 3, characterized by a flatter peak and thinner tails, indicating fewer extreme values. An example is a uniform distribution, where data points are more evenly spread. Understanding kurtosis is essential in statistical analysis as it helps in identifying the presence of outliers and the nature of data distribution, which is crucial for decision-making in fields like finance, risk management, and machine learning.
