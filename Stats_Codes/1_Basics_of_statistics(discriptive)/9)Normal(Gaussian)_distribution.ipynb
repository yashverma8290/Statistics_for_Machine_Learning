# Example
This example demonstrates the Normal (Gaussian) Distribution, which is widely used in statistics and machine learning to model real-world data. Here, we simulate a dataset of 1000 values that follow a normal distribution with a mean of 50 and a standard deviation of 10. The generated data represents a real-world scenario where values naturally cluster around the mean, with some variation. For instance, this could model students' test scores, heights of individuals, or measurement errors in an experiment. After generating the data, we calculate key statistical properties: mean (expected value), variance (measure of spread), and standard deviation (square root of variance, indicating data dispersion around the mean). The computed statistics are printed to verify that they align with the theoretical values.

To visualize the distribution, we create a histogram using Seaborn, which shows the frequency of different values in the dataset. Additionally, a Kernel Density Estimation (KDE) curve is overlaid to provide a smooth estimate of the probability density function (PDF). To compare the simulated data with theoretical expectations, we also plot the ideal normal distribution curve using the probability density function (PDF) of the normal distribution. The red line represents this theoretical curve, which closely follows the histogram, confirming that the data is normally distributed.

This analysis is useful in statistical inference, hypothesis testing, and machine learning, where normally distributed data assumptions are common. For example, it helps in grading systems, where student performance is often assumed to follow a normal distribution, or in financial modeling, where stock returns are analyzed under normal distribution assumptions. The visualization helps in understanding data spread, central tendency, and deviation, making it a valuable tool in data science and analytics.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Generate normally distributed data with mean=50, std=10
mean = 50
std_dev = 10
size = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Calculate statistics
calculated_mean = np.mean(data)
calculated_variance = np.var(data)  # Variance
calculated_std_dev = np.std(data)   # Standard Deviation

# Print results
print(f"Mean: {calculated_mean:.2f}")
print(f"Variance: {calculated_variance:.2f}")
print(f"Standard Deviation: {calculated_std_dev:.2f}")

# Plot histogram and KDE
plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, color="blue", stat="density")

# Plot theoretical normal curve
x = np.linspace(min(data), max(data), 100)
pdf_values = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf_values, 'r-', lw=2, label="Theoretical Normal Curve")

# Labels and title
plt.title("Normal Distribution with Variance & Standard Deviation")
plt.xlabel("Values")
plt.ylabel("Density")
plt.legend()
plt.show()
```

# Understanding Standard Deviation (SD) and Variance in the Graph

Standard deviation (SD) is a measure of how spread out the data points are around the mean. In this case, the SD is 10.15, which means that most of the data points lie within 10.15 units of the mean value, which is 49.94. A smaller SD would indicate that the data points are tightly clustered around the mean, while a larger SD would suggest that the values are more spread out. This helps in understanding the distribution of data and how much variation exists in the dataset.

A normal distribution follows the 68-95-99.7 rule, meaning that approximately 68% of the data falls within one SD (39.79 to 60.09) from the mean, 95% within two SDs (29.64 to 70.24), and 99.7% within three SDs (19.49 to 80.39). This means that most values in the dataset are concentrated near the mean, but some values are farther away, forming a bell-shaped curve.

Variance, which is 102.94 in this case, is simply the square of the standard deviation. It provides another way to measure the dispersion of data. A higher variance means that the data points are widely spread out, whereas a lower variance means the data points are closely packed around the mean. Since variance is measured in squared units, it is harder to interpret directly, which is why standard deviation is often preferred for practical understanding.

Looking at the graph, the blue histogram represents the actual distribution of the dataset, and the blue KDE curve shows the estimated probability density. The red curve, which represents the theoretical normal distribution, aligns well with the histogram, confirming that the dataset follows a normal distribution. The width of the bell curve is influenced by standard deviationa higher SD results in a wider and flatter curve, while a lower SD makes the curve steeper and narrower.

Since the SD is 10.15, it suggests that the data is moderately spread out around the mean. The red normal curve fits the histogram well, indicating that the data follows a normal distribution pattern. Understanding SD and variance is crucial in statistics, machine learning, and AI, as they help in analyzing data, detecting outliers, and making probabilistic predictions. Many machine learning algorithms assume normally distributed data, making these concepts essential for model accuracy and performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mu = 50  
sigma = 10  

# Generate x values
x = np.linspace(20, 80, 1000)
y = norm.pdf(x, mu, sigma)

# Plot the normal distribution
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'r-', label="Normal Distribution")
plt.fill_between(x, y, alpha=0.3)  # Shaded area
plt.title("Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x) (Probability Density)")
plt.legend()
plt.show()
```
