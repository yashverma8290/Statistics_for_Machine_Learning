# Example
This code simulates and visualizes a Binomial Distribution by modeling the probability of obtaining a certain number of successes in multiple independent trials. Here, the experiment represents flipping a fair coin 10 times per trial (with each flip having a 50% chance of landing heads). The process is repeated 1000 times to observe how often different numbers of heads appear.

First, the code defines n = 10 (number of trials per experiment) and p = 0.5 (probability of getting heads in a single flip). It then generates 1000 random samples from a binomial distribution using np.random.binomial(n, p, 1000). This simulates 1000 instances of flipping a fair coin 10 times and recording how many heads appear in each trial.

Next, the code plots a histogram using sns.histplot(), showing how frequently each possible number of heads (0 to 10) occurred in the simulation. The x-axis represents the number of heads obtained in each 10-flip trial, while the y-axis represents the probability of those occurrences.

Additionally, the code calculates the theoretical Binomial Probability Mass Function (PMF) using binom.pmf(). The PMF gives the exact probability of getting k heads in 10 flips, which is plotted as a red line to compare against the simulated data. Finally, the graph is labeled with a title, x-axis, and y-axis descriptions to clearly represent the Binomial Distribution (n=10, p=0.5).

This code helps visualize the binomial probability distribution and demonstrates how real-world simulations align with theoretical probability distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Define parameters
n = 10   # Number of trials (coin flips)
p = 0.5  # Probability of success (getting heads)

# Generate 1000 random samples from Binomial distribution
data = np.random.binomial(n, p, 1000)

# Plot histogram
plt.figure(figsize=(8,5))
sns.histplot(data, bins=range(n+2), kde=False, color="blue", stat="probability", edgecolor="black")

# Theoretical binomial probabilities
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)
plt.plot(x, y, "ro-", label="Theoretical Binomial PMF")

plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters
n_users = 1000   # Number of users who see the ad
p_click = 0.1    # Probability of clicking the ad

# Simulate ad clicks (1 = clicked, 0 = not clicked)
click_data = np.random.binomial(n=1, p=p_click, size=n_users)

# Count number of clicks (sum of ones)
total_clicks = np.sum(click_data)
print(f"Total Clicks: {total_clicks} out of {n_users}")

# Visualizing the Click Distribution
plt.figure(figsize=(8, 5))
sns.histplot(click_data, bins=2, discrete=True, kde=False)
plt.xticks([0, 1], labels=["Not Clicked", "Clicked"])
plt.xlabel("User Action")
plt.ylabel("Count")
plt.title("Ad Click Simulation using Binomial Distribution")
plt.show()
```
