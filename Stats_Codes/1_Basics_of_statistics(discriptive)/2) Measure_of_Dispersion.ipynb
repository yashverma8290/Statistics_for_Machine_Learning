```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stats

# Sample Data: Test Scores of Students
data = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100, 65, 70, 75, 80, 85, 90, 95, 60, 62, 88]

# Calculating Measures of Dispersion
range_value = max(data) - min(data)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
variance_value = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)  # Sample standard deviation

# Print Values
print("Range:", range_value)
print("Interquartile Range (IQR):", IQR)
print("Variance:", variance_value)
print("Standard Deviation:", std_dev)

# Create a figure with subplots
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Histogram for Data Distribution
sns.histplot(data, bins=10, kde=True, ax=ax[0], color="blue")
ax[0].axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label='Mean')
ax[0].axvline(np.median(data), color='green', linestyle='dashed', linewidth=2, label='Median')
ax[0].legend()
ax[0].set_title("Histogram of Data with Mean & Median")
ax[0].set_xlabel("Value")
ax[0].set_ylabel("Frequency")

# Boxplot to Show Range, IQR, and Outliers
sns.boxplot(data=data, ax=ax[1], color="orange")
ax[1].set_title("Boxplot of Data (IQR & Outliers)")
ax[1].set_ylabel("Value")

# Show plots
plt.show()
```
