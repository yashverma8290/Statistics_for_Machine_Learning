# Example
This example is based on a call center scenario where customer support calls arrive randomly. The Poisson distribution is used to model the number of calls received per minute, assuming that calls arrive independently at a constant average rate of 5 calls per minute (λ = 5). The analysis helps answer key questions related to call volume. First, it calculates the probability of receiving exactly 3 calls in a minute, which is found to be 0.1404 (14.04%), meaning that in 14.04% of the minutes, exactly 3 calls are expected. Next, it determines the probability of receiving more than 5 calls in a minute, which is 0.3840 (38.40%), indicating a 38.4% chance of exceeding 5 calls in any given minute. To further understand real-world behavior, 1000 minutes of call data are simulated based on the Poisson distribution, and a histogram is plotted to visualize the distribution of calls per minute. The red dashed line represents the mean (λ = 5), showing that most call counts are centered around 5, but fluctuations occur. This analysis has practical applications in workforce planning, where call centers can determine the number of agents needed to handle varying call volumes. It is also useful for system load balancing, allowing businesses to predict peak times and allocate resources accordingly. Additionally, it aids in queue management, helping to estimate wait times and optimize staffing levels to minimize customer dissatisfaction. This demonstrates how the Poisson distribution is valuable in modeling random events in real-world business operations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Given: Average number of calls per minute
lambda_ = 5  

# 1. Probability of exactly 3 calls arriving in a minute
prob_3_calls = poisson.pmf(3, lambda_)
print(f"Probability of exactly 3 calls in a minute: {prob_3_calls:.4f}")

# 2. Probability of more than 5 calls in a minute
prob_more_than_5 = 1 - poisson.cdf(5, lambda_)
print(f"Probability of more than 5 calls in a minute: {prob_more_than_5:.4f}")

# 3. Simulating call arrivals over 1000 minutes
simulated_calls = poisson.rvs(mu=lambda_, size=1000)

# Plot histogram of simulated data
plt.figure(figsize=(10, 5))
sns.histplot(simulated_calls, bins=15, kde=False, color="blue", stat="probability")
plt.axvline(x=5, color='r', linestyle='dashed', label="Mean (λ=5)")
plt.xlabel("Number of Calls per Minute")
plt.ylabel("Probability")
plt.title("Poisson Distribution of Call Arrivals (λ=5)")
plt.legend()
plt.show()
```

# Explanation
This histogram represents a Poisson distribution of call arrivals per minute with a mean arrival rate of λ = 5, modeling the number of customer support calls received per minute. The blue bars in the histogram show the simulated relative frequency of different numbers of calls per minute based on 1000 randomly generated samples from the Poisson distribution. The x-axis represents the number of calls per minute (ranging from 0 to 14), while the y-axis represents the probability of those occurrences. A red dashed vertical line at λ = 5 highlights the expected number of calls per minute, showing that most values are centered around 5.

At the top, two probability values provide specific insights: the probability of receiving exactly 3 calls in a minute is 0.1404 (14.04%), while the probability of receiving more than 5 calls in a minute is 0.3840 (38.40%). This aligns with the nature of the Poisson distribution, which models random, rare events occurring over a fixed time interval. The distribution is slightly right-skewed, indicating that while lower values are more frequent, higher values still have a probability of occurring. Additionally, in a Poisson distribution, the mean λ is also the variance, meaning that the spread of the data around λ = 5 is expected.

Overall, this visualization demonstrates how call arrivals follow a Poisson process, helping businesses predict and manage call volumes efficiently. By understanding this distribution, call centers can allocate resources effectively, ensuring they have enough support agents available to handle customer inquiries based on expected fluctuations in call traffic.
